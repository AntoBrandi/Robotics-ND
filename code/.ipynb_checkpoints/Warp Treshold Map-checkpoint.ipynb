From a raw image to navigable terrain in rover-centric coordinates.  Note: the images in the lower left and lower right represent the same data after a change of coordinate system.

```python
def perspect_transform(img, src, dst):

    # Get transform matrix using cv2.getPerspectivTransform()
    M = cv2.getPerspectiveTransform(src, dst)
    # Warp image using cv2.warpPerspective()
    # keep same size as input image
    warped = cv2.warpPerspective(img, M, (img.shape[1], img.shape[0]))
    # Return the result
    return warped
```

Function that performs the color treshold filtering

```python
def color_thresh(img, rgb_thresh=(0, 0, 0)):
    # Create an empty array the same size in x and y as the image 
    # but just a single channel
    color_select = np.zeros_like(img[:,:,0])
    # Apply the thresholds for RGB and assign 1's 
    # where threshold was exceeded
    for i, x in enumerate(color_select):
        for j, y in enumerate(x):
            # Treshold is lower in all the images
            if img[i,j,0]>rgb_thresh[0] and img[i,j,1]>rgb_thresh[1] and img[i,j,2]>rgb_thresh[2]:
                color_select[i,j] = 1
    # Return the single-channel binary image
    return color_select
```

Main Code that performs:
    - Open the image with matplotlib in a numpy array
    - Warp the image to obtain a planar view of the area
    - Filter the warped map with a color treshold tecnique
    - Converts into rover centric coordinates
    


Import the raw image

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt


%matplotlib inline

# Define the filename, read and plot the image
filename = 'robocam.jpg'
image = mpimg.imread(filename)
plt.imshow(image)
plt.show()
```

Warp the image to a planar image

```python
dst_size = 5
bottom_offset = 6
source = np.float32([[15,140], [300,140],[200,95], [120,95]])
destination = np.float32([[image.shape[1]/2 - dst_size, image.shape[0] - bottom_offset],
                  [image.shape[1]/2 + dst_size, image.shape[0] - bottom_offset],
                  [image.shape[1]/2 + dst_size, image.shape[0] - 2*dst_size - bottom_offset], 
                  [image.shape[1]/2 - dst_size, image.shape[0] - 2*dst_size - bottom_offset],
                  ])     

warped = perspect_transform(image, source, destination)
%matplotlib notebook
plt.imshow(warped)
plt.show() 
```

Apply the color treshold

```python
red_threshold = 160
green_threshold = 160
blue_threshold = 160

rgb_threshold = (red_threshold, green_threshold, blue_threshold)

# pixels below the thresholds
colorsel = color_thresh(warped, rgb_thresh=rgb_threshold)
%matplotlib notebook
plt.imshow(colorsel, cmap='gray')
plt.show() 
```

Convert into Rover-Centric coordinates

```python
def rover_coords(binary_img):
    # this function calculate pixel positions with reference to the rover 
    # position being at the center bottom of the image.  
    ypos, xpos = binary_img.nonzero()
    print(binary_img.shape[0])
    print(binary_img.shape[1]/2)
    x_pixel = -(ypos - binary_img.shape[0]).astype(np.float)
    y_pixel = -(xpos - binary_img.shape[1]/2 ).astype(np.float)
    return x_pixel, y_pixel
```

```python
# Extract x and y positions of navigable terrain pixels
# and convert to rover coordinates
xpix, ypix = rover_coords(colorsel)

# Plot the map in rover-centric coords
%matplotlib notebook
plt.plot(xpix, ypix, '.')
plt.ylim(-160, 160)
plt.xlim(0, 160)
plt.show() 
```
